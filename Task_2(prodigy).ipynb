```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('titanic dataset.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

**Remove null values**

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

**Exploratory Data Analysis**

Explore relationships between variables and identify patterns and trends in the data using visualizations.

**Summary Statistics**

```python
data.describe()
```

**Visualize Survival Rate by Gender**

```python
# Survival rate by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

**Visualize Survival Rate by Passenger Class**

```python
# Survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

**Visualize Age Distribution
python**

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Visualize Survival Rate by Age**


```python
# Survival rate by age
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', hue='Survived', multiple='stack', bins=30, kde=True)
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Visualize Survival Rate by Fare**




```python
# Survival rate by fare
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Fare', hue='Survived', multiple='stack', bins=30, kde=True)
plt.title('Survival Rate by Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: make a facet grid map for pclass, survived and sex

g = sns.FacetGrid(data, col='Pclass', row='Survived', height=3.2, aspect=1.6)
g.map(plt.hist, 'Sex', alpha=.5, bins=20)
g.add_legend()
```

```python
numeric_cols = data.select_dtypes(include=np.number).columns
correlation_df = data[numeric_cols].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_df, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show
```
