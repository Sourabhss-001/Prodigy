```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# prompt: reading the dataset

# Read the dataset
df = pd.read_csv('stroke dataset.csv')

# Print the first five rows of the dataset
df.head()
```

```python
# checking the null values
data_cleaned = df.isnull().sum()
data_cleaned
```

```python
# remove null values in above dataset

df.dropna(inplace=True)
```

```python
# check the null values removed or not

df.isnull().sum().sum()
```

```python
# prompt: create a gender countplot

sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

```python
# create hist plot for this

# Histogram for age distribution
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable

# Histogram for work_type distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['work_type'], bins=5, color='red', edgecolor='black')
plt.title('Work Type Distribution')
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.show()
```
